```python
# IQR method
Q1 = df1[target].quantile(0.25)
Q3 = df1[target].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter dataset
df_clean = df1[(df1[target] >= lower_bound) & (df1[target] <= upper_bound)]
print("Original rows:", len(df1), "Cleaned rows:", len(df_clean))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn
from catboost import CatBoostRegressor, Pool
from sklearn.metrics import mean_squared_error

# 5 Detect categorical columns (none here, all numeric)
cat_cols = [c for c in X_train.columns if X_train[c].dtype=='object']

# 6 CatBoost model
train_pool = Pool(X_train, y_train, cat_features=cat_cols)
val_pool = Pool(X_val, y_val, cat_features=cat_cols)

model = CatBoostRegressor(
    iterations=10000,
    learning_rate=0.05,
    depth=8,
    eval_metric="RMSE",
    random_seed=42,
    verbose=200,
    early_stopping_rounds=100
)

# 7 Train
model.fit(train_pool, eval_set=val_pool, use_best_model=True)

# 8 Validation prediction & RMSE
val_preds = model.predict(X_val)
rmse = np.sqrt(mean_squared_error(y_val, val_preds))
print("Validation RMSE:", rmse)

# 9 Actual vs Predicted plot
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_val, y=val_preds, alpha=0.5)
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'r--')  # 45 line
plt.xlabel("Actual BeatsPerMinute")
plt.ylabel("Predicted BeatsPerMinute")
plt.title("Actual vs Predicted BeatsPerMinute")
plt.show()
```
